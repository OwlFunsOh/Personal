```python
import pandas as pd
import math
from scipy import stats

dataFrame = pd.read_csv('/content/Hospital_Provider_Cost_Report_2021.csv')
print(dataFrame)

#print(dataFrame.describe())
```

The next part will collect all the rows that has a state code of 'MD'

```python
marylandData = dataFrame[dataFrame['State Code'] == 'MD']

print(marylandData)
```

Now that we have all the Maryland Hospitals, I want to see the average number of beds that a Maryland Hospital has

```python
averageMDBeds = marylandData['Number of Beds']

print(averageMDBeds.describe())
```

Sinai Hospital has 426 beds. I want to compare its capacity to other hospitals in Maryland

```python
def normalization(sampleSize, sampleStd, sampleMean, givenValue):
  score = (givenValue - sampleMean) / (sampleStd / math.sqrt(sampleSize))
  return score

sinaiZScore = normalization(59, 188.541447, 211.677966, 426)

sinaiPercentile = stats.norm.cdf(sinaiZScore) * 100

print(sinaiPercentile)
```

Sinai Hospital is very big... but it isn't the biggest in MD. What rank does it have in MD based on the number of beds? I'll find out below

```python
print(marylandData.sort_values(by = 'Number of Beds', ascending = False))
```

Counting down the list, Sinai is the 5th Largest Hospital in Maryland making it in the (1 - (5/59)) * 100 = 91.5th percentile.
