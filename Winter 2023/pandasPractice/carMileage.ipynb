```python
import pandas as pd
import matplotlib.pyplot as plt


vehicleDataFrame = pd.read_csv('/content/vehicles.csv')

compactCars = vehicleDataFrame[vehicleDataFrame['VClass'] == 'Compact Cars']

print(compactCars)
```

Now that I have loaded the Vehicle dataset, I want to plot the year and the miles per gallon on each vehicle

```python
averageCityMPG = vehicleDataFrame['city08']
year = vehicleDataFrame['year']

plt.figure()

plt.scatter(year, averageCityMPG, s = 60, alpha = 0.5, c = 'blue', edgecolors = 'black')
plt.xlabel("Model Year", fontsize = 12)
plt.ylabel("MPG", fontsize = 12)
plt.title("Fuel Efficiency of Vehicles over time")
plt.grid(True)
plt.show()
```

I don't believe there are cars nearing 160 mpg in the city.. So, I'm going to recheck the vehicle dataset and sort the cityu8 column (the city MPG column) from greatest to least

```python
sortedMPG = vehicleDataFrame.sort_values(by = 'city08', ascending = False)
print(sortedMPG[['city08', 'year', 'VClass', 'make', 'model']].iloc[0])
```

Ok... so maybe there are cars that can do that in the city. I just googled the Inoiq 6 car and verified that the hybrid car does do 153. That's insane. Now back on track, I want to get the average mpg per year and graph it.

```python
averageMPGPerYear = vehicleDataFrame.groupby("year")["city08"].mean()
print(averageMPGPerYear)
plt.figure()
plt.plot(averageMPGPerYear.index, averageMPGPerYear.values, marker = 'p', color = 'green')
plt.xlabel("Model Year", fontsize = 12)
plt.ylabel("Average MPG", fontsize = 12)
plt.title("Average MPG of vehicles over time", fontsize = 14)
plt.grid(True)
plt.show()
```
